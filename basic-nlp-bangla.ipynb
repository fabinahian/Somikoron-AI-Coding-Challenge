# 🔖 Stemming vs Lemmatization

Both Stemming and Lemmatization share the same goal, which is to bring words to their base forms.
In everyday use, words take on multiple forms according to the demands of Grammar. For example, 
> I eat, he eats, she has eaten - these three sentences represent different faces of the word 'eat'.

While both Stemming and Lemmatization tries to find the base form of words, they do the task differently. Here are some key differences between the two of them:

**Stemming**

1. Stemming works by removing the last few characters of words without any regard to context.
2. It has a higher error rate since it returns whatever is left after the reduction of affixes, even if the remaining bit isn't a word.
3. Example: 
        a. walking --> walk
        b. caring --> car
        c. stripes --> strip
        
**Lemmatization**

1. Lemmatization works by taking contexts into consideration and using a dictionary to get the base form of words.
2. It has a lower error rate because it always returns dictionary words as output.
3. Example:
        a. walking --> walk
        b. caring --> care
        c. stipes (noun) --> stripe
           stripes (verb) --> strip

# 🔖 Creating a Bag of Words

````python
#cleaning the data

clean_paragraph = bn.clean_text(paragraph)

def remove_punctuations(my_str):
    # define punctuation
    punctuations = '''````£|¢|Ñ+-*/=EROero৳০১২৩৪৫৬৭৮৯012–34567•89।!()-[]{};:'"“\’,<>./?@#$%^&*_~‘—॥”‰🤣⚽️✌�￰৷￰'''
    
    no_punct = ""
    for char in my_str:
        if char not in punctuations:
            no_punct = no_punct + char

    # display the unpunctuated string
    return no_punct

clean_paragraph = remove_punctuations(clean_paragraph)

def replace_strings(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           u"\u00C0-\u017F"          #latin
                           u"\u2000-\u206F"          #generalPunctuations
                               
                           "]+", flags=re.UNICODE)
    english_pattern=re.compile('[a-zA-Z0-9]+', flags=re.I)
    #latin_pattern=re.compile('[A-Za-z\u00C0-\u00D6\u00D8-\u00f6\u00f8-\u00ff\s]*',)
    
    text=emoji_pattern.sub(r'', text)
    text=english_pattern.sub(r'', text)

    return text

clean_paragraph = replace_strings(clean_paragraph)

clean_paragraph
````
